```python
def get_available_numbers(n):
    unique_numbers = set()
    
    pbar = tqdm(total=n)
    
    while True:
        r = requests.get('https://ocapi.onecall.no/orderflow/numbers?maxcount=100',
                         headers=headers)
        
        if r.status_code != 200:
            return
        
        numbers = r.json()['numbers']
        pbar.update(len(set(numbers).difference(unique_numbers)))
        unique_numbers.update(numbers)
        
        sleep(random.uniform(1.0, 2.0))
        
        if len(unique_numbers) >= n:
            pbar.close()
            return random.sample(unique_numbers, n)
```

```python
numbers = get_available_numbers(100)
```

```python
with open('numbers.txt', 'w') as file:
    file.write(', '.join(numbers))
```

```python
from collections import Counter
import re

print(f'Received {len(numbers)} numbers')
for n in sorted(numbers):
    one_letter = Counter(str(n)).most_common()
    two_letter = [m.group() for m in re.finditer(r'(?:(\w)\1+){1}', str(n))]
    print(f'{n}, {one_letter[:3]}, {two_letter}')
```
